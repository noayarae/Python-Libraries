#### Tunning using hyperopt
Tutorial: https://naingthet.github.io/posts/house-prices/house-prices/

```python
### ---------------- Optimization of hyperparameters ---------------
from sklearn.model_selection import cross_val_score
from hyperopt import hp, fmin, tpe, Trials, STATUS_OK
from hyperopt.pyll import scope

X_train, y_train = data[:, :-1], data[:, -1]

#E: Details of parameters here: https://xgboost.readthedocs.io/en/latest/parameter.html
space = {
    'booster': hp.choice('booster', ['gbtree']),
    'objective':  hp.choice('objective', ['reg:squarederror']),#default=reg:squarederror
    'n_estimators' : scope.int(hp.quniform('n_estimators', 100, 801, 100)),#'n_estimators' : scope.int(hp.quniform('n_estimators', 100, 1000, 1)),
    'max_depth' : scope.int(hp.quniform('max_depth', 3, 12, 1)),#'max_depth': scope.int(hp.quniform('max_depth', 2, 20, 1)), # default=6
    'gamma' : hp.choice('gamma', [0.01,0.1,0.5,1]),#'gamma': hp.uniform ('gamma', 1,9)
    'learning_rate': hp.loguniform('learning_rate',np.log(0.01),np.log(1)),# (name: eta) default=0.3 (range:[0,1]), alias: learning_rate, . 
    'subsample': hp.uniform('subsample', 0.8, 1.0), #default=1 (range:[0,1])
    #'colsample_bytree' : hp.choice('colsample_bytree', np.arange(0.3, 0.8, 0.1)),# [default=1]. Range: (0,1](Never zero). Specify the fraction of columns to be subsampled.
    'alpha' : hp.choice('alpha', np.arange(0.0, 1.1, 0.1)), # alpha [default=0, alias: reg_alpha]. L1 regularization term on weights. Increasing this value will make model more conservative.
    'criterion': hp.choice('dtree_criterion', ['gini', 'entropy'])
    }

# Running the optimization function and storing the model and trial log
# Timing
start = time.time()

# Creating an objective function that will output rmse loss given a set of model parameters
# Hyperopt will seek to minimize the loss returned by this function
def objective_function(params):
        model = XGBRegressor(**params)
        #score = cross_val_score(model, X_train_data, y_train_data, cv=5, scoring='neg_mean_squared_error')
        #score = cross_val_score(model, X_train_rfecv, y_train, cv=5, scoring='neg_mean_squared_error')
        score_mse = cross_val_score(model, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
        return {'loss':np.sqrt(-score_mse).mean(), 'status':STATUS_OK}

# Trials object will store information for each trial, allowing us to see under the hood
trials = Trials()
num_eval=30
# The fmin function will carry out the hyperopt optimization for us
# Using TPE and given the objective function, the algorithm will find the hyperparameters that minimize loss as defined by our objective function
best_param = fmin(objective_function,
                      space =space,#param_hyperopt,#space = space,#param_space, 
                      algo=tpe.suggest,
                      max_evals= num_eval,#10,#
                      trials=trials,
                      rstate= np.random.RandomState(0))

```

```python
# Loss for each trial
loss = [x['result']['loss'] for x in trials.trials]
print ("All loss:")
print (loss)
# Extract our best parameter values into a list
best_param_values = [x for x in best_param.values()]
print ("Best params")
print (best_param)
print (best_param_values)
print("Score for best parameters: ", min(loss))
print("Time elapsed: ", time.time() - start)
print("Parameter combinations evaluated: ", num_eval)
```

```python
from sklearn.model_selection import cross_validate, ShuffleSplit, cross_val_score, GridSearchCV
# Defining parameters through which grid search will iterate
# For details of parameters see: https://xgboost.readthedocs.io/en/latest/parameter.html
param_grid = [{
               'booster': ['gbtree'], #default= gbtree
               'objective' : ['reg:squarederror'], #default=reg:squarederror
               'n_estimators' : list(range(100, 501, 100)),
               'max_depth' : list(range(3, 11)), # default=6
               'gamma' : [0.01, 0.1, 0.5, 1.0] # default=0 (range:[0,∞]). Larger gamma -> more conservative the algorithm will be.
              }]

# Implement grid search using cross validation
cv_split = ShuffleSplit(n_splits=5, test_size=0.3, random_state=0)
grid_search = GridSearchCV(estimator=XGBRegressor(), 
                           param_grid=param_grid,
                           scoring = 'neg_mean_squared_error',
                           cv = cv_split)
```
